```python
sns.relplot(x='total_bill', y='tip', col='time', hue='smoker', style='smoker',
    size='size', data=tips)
plt.show()
```

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 30)
df = pd.DataFrame(data, columns=['x', 'y'])
df
```

```python
sns.jointplot(x='x', y='y', data=df);
```

```python
x, y = np.random.multivariate_normal(mean, cov, 1000).T
with sns.axes_style('white'):
    sns.jointplot(x=x, y=y, kind='hex', color='b');
```

```python
sns.jointplot(x='x', y='y', data=df, kind='kde');
```

```python
sns.set(color_codes=True, font_scale=1.5)
tp = sns.load_dataset('tips')
ax = sns.regplot(x='total_bill', y='tip', data=tp)
```

```python
np.random.seed(112)
mean, cov = [2, 3], [(1.5, 0.6), (0.6, 1)]
x, y = np.random.multivariate_normal(mean, cov, 30).T
ax = sns.regplot(x=x, y=y, color='g')
```

```python
ax = sns.regplot(x=x, y=y, ci=68)
```

```python
ax = sns.regplot(x='size', y='total_bill', data=tp, x_jitter=0.1)
```

```python
ax = sns.regplot(x='size', y='total_bill', data=tp, x_estimator=np.mean)
```

```python
tp['big_tip'] = (tp.tip/tp.total_bill) > 0.175
ax = sns.regplot( x='total_bill', y='big_tip', data=tp, logistic=True, n_boot=500,
    y_jitter=.03)
```

```python
ax = sns.regplot( x='size', y='total_bill', data=tp, x_estimator=np.mean, logx=True,
    truncate=True)
```
